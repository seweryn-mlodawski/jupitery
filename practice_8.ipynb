```python
import pandas as pd

temperatures = pd.Series([22.1, 23.4, 21.8, 24.0, 22.5], index=['Washington', 'Oregon', 'California', 'Nevada', 'Idaho'])
print("Moje temperatury:")
print(temperatures)
```

```python
#dostęp do danych 
print(f"\nTemperatura w Oregonie:", temperatures['Oregon'], "stopni")
```

```python
# operacje wektorowe
# zwiększenie wszystkich temperatur o 1 stopień
temperatures_plus_one = temperatures + 1
print("\nTemperatury po zwiększeniu o 1 stopień:")
print(temperatures_plus_one)
```

```python
#średnia temperatura
average_temperature = temperatures.mean()
print(f"\nŚrednia temperatura: {average_temperature:.2f} stopni")
```

```python
spends = pd.Series([29.99, 45.00, 15.00, 60.00, 10.00],
index=['Netflix', 'Spotify', 'iCloud', 'HBO Max', 'YouTube Premium'])
print("\nDane do zadania:")
print(spends)
```

```python
price_hbo = spends.loc['HBO Max']
print(f"\nCena subskrypcji HBO Max: {price_hbo} USD miesięcznie")
```

```python
# wyświetlenie wszystkich subskrypcji powyżej 20 USD
high_spends = spends[spends > 20]
print("\nSubskrypcje powyżej 20 USD:")
print(high_spends)
```

```python
#wyswietl subskrypcje po nazwie
print("\nCena subskrypcji HBO Max:", spends['HBO Max'], "USD miesięcznie")
```

```python
#Zajebiste jestem zajarany tym :)
#wyświetl po numerze indeksu
print("\nCena subskrypcji Netflix:", spends[0], "USD miesięcznie")
print("\nCena subskrypcji Spotify:", spends[1], "USD miesięcznie")
```

```python
# poprawka uzyj .loc i .iloc
print("\nCena subskrypcji YouTube Premium:", spends.loc['YouTube Premium'], "USD miesięcznie")
print("\nCena subskrypcji iCloud:", spends.iloc[2], "USD miesięcznie")
```

```python
data = {
    'Service': ['Netflix', 'Spotify', 'iCloud', 'HBO Max', 'YouTube Premium'],
    'Monthly Cost (USD)': [29.99, 45.00, 15.00, 60.00, 10.00],
    'Category': ['Entertainment', 'Music', 'Cloud Storage', 'Entertainment', 'Video']
}
#tworzenie DataFrame
df = pd.DataFrame(data)
print("\nDataFrame z danymi o subskrypcjach:")
display(df) #użycie display do ładniejszego wyświetlenia
```

```python
#wyciąganie pojedynczej kolumny
monthly_costs = df['Monthly Cost (USD)']
print("\nMiesięczne koszty subskrypcji:")
print(monthly_costs)
```

```python
print(df.iloc[4])
```

```python
#sprawdzenie wymiarow tabeli
print("\nWymiary DataFrame (wiersze, kolumny):", df.shape)
```

```python
# czy jest drogo ? (>30 USD)
expensive = df['Monthly Cost (USD)'] > 30
print("\nCzy subskrypcje są drogie (>30 USD)?")
print(expensive)
```

```python
#inaczej
wynik = df[df['Monthly Cost (USD)'] > 30]
print("\nSubskrypcje droższe niż 30 USD:")
display(wynik)
```

```python
# dodanie nowego wiersza
df.loc[len(df)] = ['Disney+', 12.99, 'Entertainment'] #dodanie nowej subskrypcji Disney+ len(df) zwraca aktualną liczbę wierszy, więc nowy wiersz zostanie dodany na końcu
print("\nDataFrame po dodaniu nowej subskrypcji Disney+:")
display(df)
```

Tabela rośnie w miarę dodawania nowych danych. Możemy dodawać nowe kolumny i wiersze. Dlatego że sprawdza długość len za każdym razem i dopisuja na końcu, bo ma za każdym razem aktualną długość.

```python
#przetrenuję w takim razie usuwanie wierszy i kolumn :)
df = df.drop('Disney+')
print("\nDataFrame po usunięciu subskrypcji Disney+:")
display(df)
```

z tego wywala error bo szuka disneya po indeksie a indeksem jest tu 0 1 2 itd a nie nazwa usługi. Trzeba by było zrobić reset index albo coś innego. Można powiedzieć mu że index to nazwa usługi.

```python
df = df.set_index('Service') #ustawienie kolumny 'Service' jako indeksu
print("\nDataFrame po ustawieniu kolumny 'Service' jako indeksu:")
display(df)
```

```python
#zostaw wszustko co nie nazywa sie Disney+
df = df[df['Service'] != 'Disney+']
display(df)
```

```python
df = len(df)
print("\nLiczba subskrypcji w DataFrame:", df)
```

Value_count zwraca liczbę unikalnych wartości w kolumnie.

```python
category_stat = df['Category'].value_counts()
print("\n--- Liczba unikalnych wartości w kolumnie 'Category' ---")
print(category_stat)
```

```python
import pandas as pd

# 1. Definiujemy style
styles = [
    {
        'selector': 'th',
        'props': [
            ('background-color', '#4CAF50'),
            ('color', 'white'),
            ('font-family', 'Arial'),
            ('text-align', 'center'),
            ('border', '1px solid black')
        ]
    }
]

# 2. Wyświetlamy główną tabelę (używamy display(), żeby nie zniknęła!)
print("--- Główna tabela z kolorowymi nagłówkami ---")
display(df.style.set_table_styles(styles))

# 3. Obliczamy statystyki i zamieniamy je na DataFrame, żeby też je ostylować
category_stat = df[['Category', 'Service']].value_counts().to_frame(name='Liczba wystąpień')

print("\n--- Statystyki kombinacji (też ostylowane) ---")
# Nakładamy te same style na tabelę statystyk
display(category_stat.style.set_table_styles(styles))
```
