Masking

```python
arr = np.arange(0,50,5) # tworzy tablicę od 0 do 45 z krokiem 5
arr
```

```python
arr > 15 # zwraca tablicę boolowską wskazującą, które elementy spełniają warunek
```

```python
arr[arr > 15] # wybiera elementy większe od 15
```

```python
(arr<10) | (arr>20) # używa operatora bitowego OR do łączenia warunków
```

```python
arr[(arr<10) | (arr>20)] # wybiera elementy mniejsze niż 10 lub większe niż 20
```

```python
np.where(arr <= 25) # zwraca indeksy elementów mniejszych lub równych 25
```

Ćwiczenie Wyfiltruj z arr tylko te wartości, które są mniejsze od 10 lub większe bądź równe 20, ale jednocześnie różne od liczby 40. (gdzie arr = np.arange(0,50,5)).

```python
arr = np.arange(0,50,5) # tworzy tablicę od 0 do 45 z krokiem 5
arr[(arr < 10) | ((arr >= 20) & (arr != 40))] # dokładnie to co w zadaniu
```

Mamy dane dwie tablice: a i b. Stwórz program filtrujący wartości z b, które korespondują z elementami a i są większe od 100 i mniejsze od 110.

```python
a = np.array([97,101,105,111,117,125])
b = np.array(['a','e','i','o','u','y'])

mask = (a > 100) & (a < 110)
result = b[mask]
result
```

(a > 100) & (a < 110): NumPy sprawdza te warunki dla każdego elementu tablicy a.

97: Fałsz

101: Prawda (spełnia oba warunki)

105: Prawda (spełnia oba warunki)

111: Fałsz (za duże)

...i tak dalej.
W rezultacie powstaje maska: [False, True, True, False, False, False].

b[maska]: Bierzemy z tablicy b tylko te elementy, które w masce mają wartość True. Odpowiada to literom 'e' oraz 'i'.

```python
np.array([[0,4,1],[2,0,4]])
```

```python
tablica = np.array([[0,4,1],[2,0,4]])
minima = tablica.min(axis=1)  # minimalne wartości w każdym wierszu
minima_kolumn = tablica.min(axis=0)  # minimalne wartości w każdej kolumnie
maksima = tablica.max(axis=1)  # maksymalne wartości w każdym wierszu
maksima_kolumn = tablica.max(axis=0)  # maksymalne wartości w każdej kolumnie
print("Minima w wierszach:", minima)
print("Minima w kolumnach:", minima_kolumn)
print("Maksima w wierszach:", maksima)
print("Maksima w kolumnach:", maksima_kolumn)
```
