```python
macierz = np.arange(9).reshape(3,3)
print(macierz)
macierz[-1, -1]  = 999 # zmienia ostatni element na 999
macierz[1:,1:] = 0 # ustawia dolny prawy 2x2 blok na 0
print('\n', macierz)
```

Ćwiczenie - znajdź indeksy niezerowych elementów z tablicy

```python
arr  = np.array([1,2,0,0,4,0])
np.nonzero(arr) # zwraca indeksy niezerowych elementów
```

```python
arr  = np.array([1,2,0,0,4,0])
np.where(arr != 0) # zwraca indeksy elementów spełniających warunek - metoda bardziej uniwersalna niż nonzero
```

Masking

```python
arr = np.arange(0,50,5) # tworzy tablicę od 0 do 45 z krokiem 5
arr
```

```python
arr > 15 # zwraca tablicę boolowską wskazującą, które elementy spełniają warunek
```

```python
arr[arr > 15] # wybiera elementy większe od 15
```
